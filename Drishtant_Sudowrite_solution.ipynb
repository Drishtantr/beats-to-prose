## Prose Generation Coding Task

The challenge involves creating a backend system that transforms a series of bulleted "beats" into well-crafted prose that you would find in a novel. This tool is designed to aid writers in fleshing out as scene based on their outlines.

**Part 1: Beats-to-Prose**
Given a set of story beats, like this set for a sci-fi chapter:

1. Begin the chapter with Jack and Xander continuing their excavation on the lunar surface,
creating a sense of tension and anticipation.
2. Describe the barren landscape of the moon, emphasizing its desolation and the isolation
felt by Jack and Xander.
3. Use vivid and descriptive language to portray the moment when Jack and Xander uncover
an alien artifact, highlighting its mysterious and otherworldly appearance.
4. Convey Jack's excitement and trepidation as he realizes the significance of the discovery
and the potential implications it holds for humanity.
5. Show Jack immediately contacting Dr. Selene Thorne, emphasizing his trust in her and the
urgency of the situation.
6. Include dialogue between Jack and Dr. Thorne, showcasing their professional relationship
and the gravity of the situation.
7. Highlight the importance of the discovery and the potential consequences it could have on
Earth and the lunar mining project.
8. Show Jack and Xander waiting anxiously for further instructions from Dr. Thorne, creating a
sense of anticipation and uncertainty.
9. Emphasize the isolation and vastness of the lunar landscape, reinforcing the challenges
and dangers that Jack and Xander face.
10. Use sensory details to immerse the reader in the lunar environment, such as the crunch of
lunar soil beneath their boots and the cold, airless atmosphere.
11. Portray Jack's determination and resolve to protect the discovery, even in the face of
potential conflict with rival miners.
12. Hint at the potential dangers and obstacles that Jack and Xander may encounter as they
navigate through a world of corporate greed and betrayal.
13. Foreshadow the secrets and mysteries that the alien artifact holds, building intrigue and
anticipation for future chapters.
14. End the chapter with a cliffhanger or unresolved tension, leaving the reader eager to
continue reading and discover what happens next.

Use these beats to produce ~1500 words of prose that matches these beats - in other words, if
a human were to summarize the prose, they would produce something along the lines of these
beats. It's okay to hallucinate details if they are missing from the beats in order to create the
narrative. Target approximately 100-150 words of output per beat. You are allowed to use any LLM or NLP techniques, including open source and closed source services like GPT or Claude.

```python
class ProseGenerator:
    def __init__(self, api_key):
        genai.configure(api_key=API_KEY)
        self.model = genai.GenerativeModel("gemini-1.5-flash")
        self.nlp = spacy.load("en_core_web_sm")

    def generate_prose(self, beats, word_count_per_beat=150):
        """
        Transform story beats into a cohesive narrative prose using Gemini
        
        Args:
            beats (list): Sequence of story beats
            word_count_per_beat (int): Approximate words per beat
        
        Returns:
            str: Generated narrative prose
        """
        full_prompt = self._construct_narrative_prompt(beats)
        
        try:
            response = self.model.generate_content(full_prompt)
            generated_text = response.text
            # generated_text = fallback
            return self._post_process_narrative(generated_text)
        except Exception as e:
            print(f"Generation error: {e}")
            return ""
    
    def _construct_narrative_prompt(self, beats):
        """Construct a detailed prompt for narrative generation"""
        
        context = """Generate a science fiction prose narrative about lunar excavation 
        with dramatic tension, character depth, and scientific authenticity. 
        Follow these story progression beats precisely:"""
        
        beat_instructions = "\n".join([f"• {beat}" for beat in beats])
        
        return f"{context}\n{beat_instructions}\n\nEnsure the prose is immersive, scientifically plausible, and maintains narrative momentum."
    
    def _post_process_narrative(self, text):
        """
        Post-process the generated narrative for improved coherence
        
        - Use spaCy for basic linguistic refinement
        - Check narrative flow
        """
        doc = self.nlp(text)
        cleaned_text = " ".join([sent.text for sent in doc.sents])
        
        return cleaned_text

beats = [
    "Begin the chapter with Jack and Xander continuing their excavation on the lunar surface, creating a sense of tension and anticipation.",
    "Describe the barren landscape of the moon, emphasizing its desolation and the isolation felt by Jack and Xander.",
    "Use vivid and descriptive language to portray the moment when Jack and Xander uncover an alien artifact, highlighting its mysterious and otherworldly appearance.",
    "Convey Jack's excitement and trepidation as he realizes the significance of the discovery and the potential implications it holds for humanity.",
    "Show Jack immediately contacting Dr. Selene Thorne, emphasizing his trust in her and the urgency of the situation.",
    "Include dialogue between Jack and Dr. Thorne, showcasing their professional relationship and the gravity of the situation.",
    "Highlight the importance of the discovery and the potential consequences it could have on Earth and the lunar mining project.",
    "Show Jack and Xander waiting anxiously for further instructions from Dr. Thorne, creating a sense of anticipation and uncertainty.",
    "Emphasize the isolation and vastness of the lunar landscape, reinforcing the challenges and dangers that Jack and Xander face.",
    "Use sensory details to immerse the reader in the lunar environment, such as the crunch of lunar soil beneath their boots and the cold, airless atmosphere.",
    "Portray Jack's determination and resolve to protect the discovery, even in the face of potential conflict with rival miners.",
    "Hint at the potential dangers and obstacles that Jack and Xander may encounter as they navigate through a world of corporate greed and betrayal.",
    "Foreshadow the secrets and mysteries that the alien artifact holds, building intrigue and anticipation for future chapters.",
    "End the chapter with a cliffhanger or unresolved tension, leaving the reader eager to continue reading and discover what happens next."
]

generator = ProseGenerator(api_key=API_KEY)
new_narrative = generator.generate_prose(beats)
```

```python
class EnhancedProseGenerator:
    def __init__(self, api_key):
        genai.configure(api_key=api_key)
        self.model = genai.GenerativeModel("gemini-1.5-flash")
    
    def generate_prose(self, beats, characters, setting, genre, prose_style):
        """
        Generate prose with enhanced contextual parameters
        
        Args:
            beats (list): Story progression beats
            characters (list): Character metadata
            setting (dict): Story setting details
            genre (str): Story genre
            prose_style (str): Desired writing style
        """
        enhanced_prompt = self._construct_enhanced_prompt(
            beats, characters, setting, genre, prose_style
        )
        
        response = self.model.generate_content(enhanced_prompt)
        return response.text
    
    def _construct_enhanced_prompt(self, beats, characters, setting, genre, prose_style):
        """Create a comprehensive, context-rich generation prompt"""
        character_context = "\n".join([
            f"Character: {char['name']} - {char['description']}" 
            for char in characters
        ])
        
        prompt_template = f"""
        Generate a {genre} narrative with {prose_style} prose style.
        
        Story Context:
        Setting: {setting}
        Characters:
        {character_context}
        
        Story Beats:
        {" ".join(beats)}
        
        Generate a narrative that embodies the specified style and context.
        """
        
        return prompt_template
        
characters = [
    {
        "name": "Jack Martinez",
        "description": (
            "A tenacious and determined lunar excavation leader in his mid-30s, with a sharp scientific mind. "
            "Tall and lean, Jack’s sun-weathered face and dark eyes betray his countless days spent in the harsh lunar environment. "
            "His passion for exploration drives him to push the boundaries of human achievement, though his impulsive nature sometimes leads him into precarious situations. "
            "Jack often relies on his quick thinking and instinct, making him both a courageous and unpredictable figure among his crew."
        )
    },
    {
        "name": "Dr. Selene Thorne",
        "description": (
            "A brilliant and calculating mission controller in her early 40s, Dr. Selene Thorne is the calm, measured force behind the excavation. "
            "Her meticulous attention to detail and ability to stay calm under pressure have made her indispensable to the success of the mission. "
            "With short-cropped hair, glasses perched on the edge of her nose, and a tendency to always wear a faint, unreadable smile, Selene exudes an air of intellectual superiority and patience. "
            "While she values logic above all else, her hidden empathy often shows through when it comes to Jack’s reckless endeavors."
        )
    }
]

setting = {
    "location": (
        "The barren, silent expanse of the Lunar surface, in the desolate region known as Tranquility Base. "
        "The jagged, gray landscape stretches endlessly beneath the harsh, unyielding sunlight. The horizon is a distant blur, "
        "where the Earth hangs like a fragile marble in the blackness of space, a constant reminder of the isolation they face. "
        "Deep craters, rocky outcrops, and scattered remnants of past missions give the place an eerie, forgotten atmosphere, "
        "as if the moon itself is silently watching over their every move."
    ),
    "time": "Mid-22nd century, an era of great technological advances and corporate-driven space exploration, "
            "but also one marked by political unrest and a growing sense of uncertainty about humanity’s future in space.",
    "atmosphere": (
        "An isolated, tension-filled research environment, where every decision feels weighty, and every moment is fraught with danger. "
        "The airless void outside and the confined spaces of the lunar base add to the pervasive sense of claustrophobia. "
        "The constant hum of machinery is the only sound that breaks the unnerving silence, and even then, it feels like the walls themselves are watching, "
        "waiting for something to happen—something that might just alter the course of human history."
    )
}

genre = "Hard Sci-Fi / Thriller"
prose_style = "Cinematic and Suspenseful"

generator = EnhancedProseGenerator(api_key=API_KEY)
enchanced_prose = generator.generate_prose(
    beats=beats,
    characters=characters,
    setting=setting,
    genre=genre,
    prose_style=prose_style
)
print(enchanced_prose)
```
